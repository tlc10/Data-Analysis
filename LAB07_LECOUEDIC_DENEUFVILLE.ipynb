# A Text processing with NLTK

## 1

```python
metamorphosis_clean = 'metamorphosis_clean.txt'
text_open = open(metamorphosis_clean, 'rt')
text = text_open.read()
text.split('\n')[0:5]
```

```python
sentences = sent_tokenize(text)
print("Il y a {} phrases." .format(len(sentences)))
print(sentences[0])
```

```python
first_sentence = sentences[0]
words = word_tokenize(first_sentence)
print(words)
nltk.pos_tag(words)
```

nltk.pos_tag(words) retourne une liste de tuples. Pour chaque tuple, on retrouve le mot du texte en indice 0 et sa catégorie en indice 1 (adverbe, conjonction de coordination, adjectif, nom,...).Les tuples de la liste sont dans le même ordre que les mots du texte.

```python
split_text = text.split()
print(split_text[0:10])
```

Avec word_tokenize, on obtient une liste avec tous les caractères (mots, virgules, points, ...) alors qu'avec split(), on obtient une liste qu'avec les mots.

```python
stop_words = stopwords.words('english')
print(stop_words)
```

Les stopwords représentent une liste de mots non exhaustives n'ayant pas d'utilité pour le traitement du language naturel. Ce sont en général des mots d'usage courant qui ne sont pas pertinents pour réaliser du NLP.

```python
for r in split_text: 
    if r in stop_words:
        split_text.remove(r)
new_split_text=" "
new_split_text = new_split_text.join(split_text)
sentences_new_split_text = sent_tokenize(new_split_text)
print(sentences_new_split_text[0])
```

## 2

```python
doc_02 = TextBlob("Paris is so beautiful, I like Paris, people are so nice !")
```

```python
doc_02.sentiment
```

On constate que la "polarity" de cette phrase à une valeur de 0.8 ce qui est relativement élevé car proche de 1. L'algorithme a donc bien cerné que la phrase avait un caractère positif. La "subjectivity" est égale à 1, donc l'algorithme détecte que cette  phrase est une phrase décrivant une opinion personnelle. On peut dire que la prédection est plutôt correcte !

## 3

```python
wiki_text = 'wiki.txt'
wiki_text_open = open(wiki_text, 'rt')
text = wiki_text_open.read()
doc_03 = TextBlob(text)
print(doc_03)
```

```python
words_list = doc_03.words
print("Il y a {} mots." .format(len(words_list)))
```

```python
sentences_list = doc_03.sentences
print("Il y a {} phrases." .format(len(sentences_list)))
```

```python
for doc_03.sentence in doc_03.sentences : 
    print(doc_03.sentence.sentiment)
```

On remarque que la plupart des valeurs de polarité sont proches de 0. Ceci est plutôt normal car le texte n'exprime pas particulièrement de sentiments positifs ou négatifs, il y a une forme de neutralité sentimentale dans ce texte. Les valeurs de subjectivité sont aussi proches de 0. Cela signifie que l'algorithme émet l'hypothèse que ces phrases ont plutôt un caractère objectif. Ce qui est plutôt vrai car en lisant le texte, on s'aperçoit qu'il ne fait que relater des faits historiques, ce qui est très factuel.
